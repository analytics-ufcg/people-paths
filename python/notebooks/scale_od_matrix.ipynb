### Origin-Matched Boardings Imputation

```python
origin_known.head()
```

```python
origin_matched_boardings = origin_known.merge(user_trips_data, left_on=['cardNum','boarding_id','o_route'], right_on=['cardNum','o_boarding_id','o_route']) \
                                        .assign(start_hour = lambda x: x.o_boarding_datetime.dt.hour)
```

```python
origin_matched_boardings.head()
```

```python
origin_matched_boardings[pd.isna(origin_matched_boardings['o_stopPointId'])]
```

```python
od_matrix.columns
```

```python
od_imputation_basis = od_matrix.filter(['route','busCode','from_stop_id','start_time','to_stop_id','end_time']) \
                                .assign(route = lambda x: x.route.astype(str)) \
                                .assign(start_hour = lambda x: x.start_time.dt.hour)
```

```python
od_imputation_basis.head()
```

```python
od_imputation_basis.dtypes
```

```python
dest_alternatives = origin_matched_boardings.merge(od_imputation_basis, left_on=['o_route','o_stopPointId','start_hour'], right_on=['route','from_stop_id','start_hour'])
```

```python
dest_dist = dest_alternatives.groupby(['route','from_stop_id','start_hour','to_stop_id']) \
                            .agg({'cardNum':'count'}) \
                            .reset_index() \
                            .rename(index=str, columns={'cardNum':'freq'})
```

```python
dest_dist.head()
```

```python
origin_dist = dest_alternatives.groupby(['route','from_stop_id','start_hour']) \
                                .agg({'to_stop_id': 'count'}) \
                                .reset_index() \
                                .rename(index=str, columns={'to_stop_id':'total_freq'})
```

```python
origin_dist.head()
```

```python
od_dist = origin_dist.merge(dest_dist) \
            .assign(perc = lambda x: x.freq/x.total_freq) \
            .filter(['route','from_stop_id','start_hour','to_stop_id','perc'])
```
