## Visualize Differences

#### Creating a Divergent Scale Palette

```python
div_scale_cmap = sns.diverging_palette(240, 0, n=10, as_cmap=True)
```

#### Baseline - Survey 

```python
baseline_survey_macrozones = baseline_macrozones - survey_macrozones
```

```python
baseline_survey_macrozones
```

```python
sns.heatmap(baseline_survey_macrozones, cmap=div_scale_cmap, center=0.0)
```

#### All Est. Trips - Survey 

```python
all_est_survey_macrozones = all_est_trips_macrozones - survey_macrozones
```

```python
all_est_survey_macrozones
```

```python
sns.heatmap(all_est_survey_macrozones, cmap=div_scale_cmap, center=0.0)
```
