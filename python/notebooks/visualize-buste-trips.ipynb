#### Plot Trips made by users

```python
locationlist = buste_sample[['gpsLat','gpsLon']].values.tolist()
locationlist = locationlist[4] + locationlist[7]
len(locationlist)
locationlist
```

```python
m = folium.Map([-25.4244287,-49.2653819], zoom_start=11)
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point]).add_to(m)
m
```

```python
# convert to (n, 2) nd-array format for heatmap
stationArr = buste_sample[['gpsLat', 'gpsLon']].as_matrix()

# plot heatmap
m.add_children(plugins.HeatMap(stationArr, radius=15))
m
```
